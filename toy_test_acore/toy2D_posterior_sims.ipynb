# Simulations with different values of the prior

#### - a = 1, b = 1

```python
data = generate_data(sample_size=10000, 
                     beta_a=1, beta_b=1, 
                     lower_theta=0, higher_theta=100, 
                     scale=5, 
                     split=False)
data.head()
```

#### $\theta$ is uniformly distributed if $\gamma \sim \text{Beta}(1,1)$, where $\theta = (\theta_{high} - \theta_{low})\gamma + \theta_{low}$

```python
sns.distplot(data.theta, kde=False)
plt.title(r"$\theta \sim Beta(1, 1)$")
plt.show()
```

#### Equal to uniform as before

```python
plot_data(data=data.to_numpy(), fig_size=(12, 10))
```

### Regression 

```python
train, test = generate_data(sample_size=10000, 
                            beta_a=1, beta_b=1, 
                            lower_theta=0, higher_theta=100, 
                            scale=5,
                            test_fraction=0.1,
                            split=True)
regression_bias(train=train, test=test, plot_title=r"$Beta(1,1)$")
```
